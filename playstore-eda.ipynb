```python
# sort the above data
sort_data = df.isnull().sum().sort_values(ascending=False)
sort_data
```

```python
# lets see percentage of missing values
missing_perc = sort_data/len(df) * 100
missing_perc
```

```python
# plot of missing/null value 
import missingno as msno
msno.matrix(df)
```

```python
missing_perc = missing_perc[missing_perc != 0]
# plot of missing/null values percentage
import matplotlib
matplotlib.rcParams['figure.figsize']= (20,6)
missing_perc.plot(kind='barh')
```

```python
df.dropna(subset=['App Name','Developer Email','Developer Id','Minimum Installs','Installs','Currency','Size'],inplace=True)

```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.nunique().sort_values(ascending=False)
```

```python
# Top 10 Categories having highest Number of Reviews
top_10_cat = df.groupby('Category')['Rating Count'].max().sort_values(ascending=False).reset_index()
top_10_cat
```

```python

sns.catplot(x='Category',y='Rating Count', data=top_10_cat.head(10),kind='bar')
plt.xticks(rotation=90)
plt.show()
```

```python
# What are the top downloaded apps in each of these categories?
top_apps = df.groupby(['App Name'])[['Maximum Installs']].sum().sort_values(by='Maximum Installs', ascending=False)
top_apps.head(10)
```

```python
top_apps.head(10).plot(kind='barh')
```
